# Data cleaning 

This project focuses on cleaning the real-world audiobook dataset scraped from Audible. This messy dataset contains multiple inconsistencies such as wrong data types, redundant texts, mixed formats, and unstructured fields.

The objective of this project is to transform the dataset into a clean and structured format Python (pandas), making it suitable for further analysis and visualization.

## 3) Data Summary and Key Issues in the data

- The dataset have 87489 rows and 8 columns.
- All columns of dataset are of object type.

#### Column Description
- `name` - audiobook title
- `author` - author name
- `narrator` - narrator name
- `time` - duration or length of the audiobook
- `releasedate` - release date of the audiobook
- `language` - language of the audiobook
- `stars` - rating of the audiobook and number of ratings
- `price` - price of the audiobook

### Key Issues in the Dataset


#### 1. Redundant Text in Columns
- The `author` and `narrator` columns contain unnecessary prefixes such as **Writtenby:** and **Narratedby:**.
- These prefixes do not add analytical value and need to be removed.

#### 2. Improper Name Formatting
- Author and narrator names are not properly formatted.There is not space between first and last name of author and narrator.
- This reduces readability and consistency.


#### 3. Incorrect Data Type (Time Column)
- The `time` column is stored as text (object type).
- It contains mixed formats like ""2 hrs and 20 mins""
- This is not suitable for any numerical analysis.


#### 4. Incorrect Date Format
- The `releasedate` column is stored as an object instead of datetime.

#### 5. Incorrect Data Type of price Column
- The `price` column is stored as text and contains commas.
- `price` columns should of float type.


#### 6. Multiple Values in stars column
- The `stars` column contain two pieces of information:
  - Rating (e.g., 4.5)
  - Number of ratings (e.g., 41 ratings)
- The column needs to be split into two sepertae columns.

#### 7. Columns are Stored as Object Type
- Every column in the dataset is of object type.
- We need to assign correct data type for these.

## 4) Data Cleaning

```python
df.head()
```

#### 1) Cleaning 'author' column

```python
# Removing 'Writtenby:' prefix from author column.
df["author"] = df["author"].str.replace("Writtenby:", "")
```

```python
# Adding space between the first and last name
df["author"] = df["author"].str.replace(r"([a-z])([A-Z])", r"\1 \2", regex=True)
```

#### 2) Cleaning 'narrator' column

```python
# Removing 'Narratedby:" prefix from narrator column.
df["narrator"] = df["narrator"].str.replace("Narratedby:", "")
```

```python
# Adding space between narrator's first and last name
df["narrator"] = df["narrator"].str.replace(r"([a-z])([A-Z])", r"\1 \2", regex=True)
```

#### 3) Cleaning 'time' column

```python
df["time"]
```

```python
# Extracting hours from time column
df["hours"] = df["time"].str.extract(r"(\d+)\s*hr[s]?").fillna(0).astype(int)
```

```python
# Extacting minutes from time column
df["minutes"] = df["time"].str.extract(r"(\d+)\s*min[s]?").fillna(0).astype(int)
```

```python
# Creating new column 'duration(mins)' 
df["duration(min)"] = df["hours"] * 60 + df["minutes"]
```

#### 4) Converting 'date' column to datetime type 

```python
df["releasedate"] = pd.to_datetime(df["releasedate"], format="%d-%m-%y")
```

#### 5) Cleaning 'stars' column

```python
# Creating new column 'ratings' from stars
df["ratings"] = df["stars"].str.extract(r"(\d(?:\.\d)?) out").astype(float)
```

```python
# Creating new columnn 'numberofratings" from stras
df["numberofratings"] = df["stars"].str.extract(r"(\d+) rating").fillna(0).astype("int64")
```

#### 6) Cleaning 'price' column

```python
# 338 records in the price column are labeled as 'Free', we can replace this with 0
df[df["price"] == "Free"].shape
```

```python
df["price"] = df["price"].str.replace("Free", "0").str.replace(",", "").astype(float)
```

```python
df.head()
```

#### 7) Droping unwanted columns and rearranging columns

```python
df = df[["name", "author", "releasedate", "time", "duration(min)",
         "language", "ratings", "numberofratings", "price"]]
```
